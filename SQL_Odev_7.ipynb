```python
import mysql.connector
```

```python
from sqlalchemy import create_engine

engine = create_engine('mysql+mysqlconnector://odev:password@localhost/sakila')
# select .. from ..  limit sayı ile dönecek kayıt sayısını 
# sınırlayabileceğimizi hatırlayalım..
connection = engine.connect()
result = connection.execute("select first_name from customer")
for row in result:
    print("Musteri ismi:", row['first_name'])
    
# bağlantıyı kapatmayı unutmamamız lazım
connection.close()
```
